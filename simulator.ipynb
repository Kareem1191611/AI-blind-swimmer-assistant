```python
import cv2
import numpy as np

# Frame size
width = 800
height = 600

def create_pool():
    frame = np.zeros((height, width, 3), dtype=np.uint8)
    
    # Blue pool background
    frame[:] = (255, 100, 0)  # BGR (blue-ish)

    return frame
```

```python
def draw_lane(frame):
    left_lane_x = 250
    right_lane_x = 550

    cv2.line(frame, (left_lane_x, 0), (left_lane_x, height), (255, 255, 255), 3)
    cv2.line(frame, (right_lane_x, 0), (right_lane_x, height), (255, 255, 255), 3)

    return left_lane_x, right_lane_x
```

```python
class Swimmer:
    def __init__(self):
        self.x = 400
        self.y = 0
        self.speed = 3

    def move(self):
        self.y += self.speed  # move forward
        if self.y == 25 :
            self.x -= 1.5

    def draw(self, frame):
        # cv2.ellipse(image, center, axes, angle, startAngle, endAngle, color, thickness)
        cv2.ellipse(frame, (int(self.x), int(self.y)),
                    (20, 20), 0, 0, 360,
                    (0, 0, 255), 11)
```

```python
def move(self):
    self.y += self.speed
    
    # simulate drift
    drift = np.random.uniform(-1, 1)
    self.x += drift
```

```python
swimmer = Swimmer()

while True:
    
    frame = create_pool()
    left_lane_x, right_lane_x = draw_lane(frame)

    swimmer.move()
    swimmer.draw(frame)

    lane_center = (left_lane_x + right_lane_x) / 2
    drift = swimmer.x - lane_center

    threshold = 100
    if swimmer.y >= height - 40:
        command = "TURN NOW"
        print(drift)
    elif drift > threshold:
        command = "Move Left"
        print(drift)
    elif drift < -threshold:
        command = "Move Right"
        print(drift)
    else:
        command = "Centered"

    cv2.putText(frame, command, (50,50),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (0,255,0), 2)

    cv2.imshow("Swimming Guidance Simulation", frame)

    if cv2.waitKey(30) == 27:
        break

cv2.destroyAllWindows()
```
